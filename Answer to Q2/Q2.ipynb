```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False
```

```python
c_S = np.arange(0, 10, 0.000001)
k_1 = 100 / 60
k_2 = 600 / 60
k_3 = 150 / 60
c_E0 = 1
v = (k_1 * k_3 * c_E0 * c_S)/(k_2 + k_3 + k_1 * c_S)

plt.xlabel('c_S')
plt.ylabel('v')
plt.title("Schematic of clasic Michaelis-Menten equation plot")
plt.plot(c_S, v)
plt.show()
```
